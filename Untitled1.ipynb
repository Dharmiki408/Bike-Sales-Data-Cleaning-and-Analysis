```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = '/content/Sales.csv'  # Updated path
# Changed pd.read_excel to pd.read_csv
df = pd.read_csv(file_path)

# Show initial info
print("Initial Data Info:")
print(df.info())
print("\nMissing values:")
print(df.isnull().sum())

# 1. Handle missing values
if 'Day' in df.columns:
    df['Day'].fillna(df['Day'].mode()[0], inplace=True)
if 'Age_Group' in df.columns:
    df['Age_Group'].fillna(df['Age_Group'].mode()[0], inplace=True)
if 'Product_Description' in df.columns:
    df['Product_Description'].fillna('Unknown', inplace=True)
if 'Order_Quantity' in df.columns:
    df['Order_Quantity'].fillna(df['Order_Quantity'].median(), inplace=True)
if 'Customer_Age' in df.columns:
    df['Customer_Age'] = df['Customer_Age'].fillna(df['Customer_Age'].median())
if 'Day' in df.columns:
    df['Day'] = df['Day'].fillna(df['Day'].mode()[0])

# 2. Remove duplicates
df.drop_duplicates(inplace=True)

# 3. Standardize text columns
for col in ['Customer_Gender', 'Country', 'State', 'Age_Group', 'Product_Category', 'Sub_Category']:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip().str.title()
        if col == 'Customer_Gender':
            df[col] = df[col].str.upper()
        if col == 'Country':
            df[col] = df[col].replace({'United  States': 'United States'})

# 4. Convert date columns
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# 5. Rename columns to lowercase and snake_case
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# 6. Fix numeric data types
if 'customer_age' in df.columns:
    df['customer_age'] = df['customer_age'].astype(int)
if 'day' in df.columns:
    df['day'] = df['day'].astype(int)
if 'order_quantity' in df.columns:
    df['order_quantity'] = df['order_quantity'].astype(int)
for col in ['unit_cost', 'unit_price', 'profit', 'cost', 'revenue']:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')

# Save cleaned dataset
df.to_csv('cleaned_bike_sales_dashboard.csv', index=False)

# Create a README summary
with open('README.md', 'w') as f:
    f.write("# Task 1 - Data Cleaning and Preprocessing\n")
    f.write("\n## Summary of Changes:\n")
    f.write("- Filled missing values using mode, median, or placeholder where appropriate.\n")
    f.write("- Removed duplicate rows.\n")
    f.write("- Standardized text fields (e.g., gender, country, category names).\n")
    f.write("- Parsed the 'Date' column to datetime format.\n")
    f.write("- Renamed columns to lowercase with underscores.\n")
    f.write("- Converted numeric fields and fixed data types.\n")
    f.write("\nCleaned dataset saved as `cleaned_bike_sales_dashboard.csv`.\n")
    display(df)# Assuming 'df' is your pandas DataFrame loaded from the previous step

import matplotlib.pyplot as plt
import seaborn as sns

# Example: Show the distribution of the 'customer_age' column

# Histogram
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='customer_age', kde=True) # kde=True adds a KDE line
plt.title('Distribution of Customer Age')
plt.xlabel('Customer Age')
plt.ylabel('Frequency')
plt.show()

# Kernel Density Estimate (KDE) plot
plt.figure(figsize=(8, 6))
sns.kdeplot(data=df, x='customer_age', fill=True) # fill=True fills the area under the curve
plt.title('KDE of Customer Age')
plt.xlabel('Customer Age')
plt.ylabel('Density')
plt.show()

# Box Plot - Useful for seeing quartiles and outliers
plt.figure(figsize=(8, 6))
sns.boxplot(data=df, y='customer_age') # Using y for a vertical box plot
plt.title('Box Plot of Customer Age')
plt.ylabel('Customer Age')
plt.show()

# Example: Show the distribution of 'order_quantity' using a histogram
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='order_quantity', bins=20) # bins controls the number of bars
plt.title('Distribution of Order Quantity')
plt.xlabel('Order Quantity')
plt.ylabel('Frequency')
plt.show()

# Example: Compare the distribution of 'customer_age' by 'customer_gender' using box plots
if 'customer_gender' in df.columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(data=df, x='customer_gender', y='customer_age')
    plt.title('Customer Age Distribution by Gender')
    plt.xlabel('Customer Gender')
    plt.ylabel('Customer Age')
    plt.show()

# Example: Show the distribution of 'profit' using a violin plot
if 'profit' in df.columns:
    plt.figure(figsize=(8, 6))
    sns.violinplot(data=df, y='profit')
    plt.title('Distribution of Profit')
    plt.ylabel('Profit')
    plt.show()
```
